```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

# INNER JOIN

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}
```

```python
right_table = {'CountryID': [3,4],
              'Country': ['Pandama', 'Spain']}
```

```python
left_table_df = pd.DataFrame(left_table)
right_table_df = pd.DataFrame(right_table)

print(left_table_df)
right_table_df
```

```python
inner_join = left_table_df.merge(right_table_df, on='CountryID')
```

```python
inner_join
```

```python
inner_join = left_table_df.merge(right_table_df, on='CountryID').drop('Units', axis=1)
inner_join
```

### Excuting by changinhg the name of right_table key name from 'CountryID' to 'ID', to explore the another way of getting inner join values

```python
right_table = {'ID': [3,4],
              'Country': ['Pandama', 'Spain']}
```

```python
right_table_df = pd.DataFrame(right_table)
right_table_df
```

```python
inner_join = left_table_df.merge(right_table_df, left_on='CountryID', right_on='ID')
inner_join
```

```python
inner_join = left_table_df.merge(right_table_df, left_on='CountryID', right_on='ID').drop('Units', axis=1)
inner_join
```

```python
inner_join = left_table_df.merge(right_table_df, left_on='CountryID', right_on='ID', suffixes=['_left','_right']).drop('Units', axis=1)
inner_join
```

# Left Anti Join

```python
left_table = {'Date': ['01-01-2020', '02-01-2020', '03-01-2020', '04-01-2020' ],
              'CountryID': [1,1,3,2],
              'Units': [40, 25, 30, 35]}

left_table_df = pd.DataFrame(left_table)
left_table_df
```

```python
right_table = {'ID': [3,4],
              'Country': ['Panama', 'Spain']}

right_table_df = pd.DataFrame(right_table)
right_table_df
```

```python
left_join= left_table_df.merge(right_table_df,left_on='CountryID',right_on='ID',how='left')
left_join
```

```python
left_anti_join = left_join[left_join['Country'].isna()][['Date', 'CountryID', 'Units']]
left_anti_join
```

```python
left_anti_join = left_join[left_join['_merge']=='left_only']#.drop(['ID','Country'],axis=1)
left_anti_join
```

# Right Anti Join

```python
right_table = {'ID': [3,4],
              'Country': ['Panama', 'Spain']}

right_table_df = pd.DataFrame(right_table)
right_table_df
```
