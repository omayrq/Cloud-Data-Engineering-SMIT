```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Creating a basic figure
# This creates a new figure with default size and settings.
plt.figure()
plt.plot([1, 2, 3, 4], [1, 4, 2, 8])
plt.title("Simple Figure")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# 2. Creating a figure with a specified size (figsize)
# figsize is a tuple (width, height) in inches.
plt.figure(figsize=(8, 6))
plt.plot(np.sin(np.linspace(0, 2 * np.pi, 100)))
plt.title("Figure with Custom Size")
plt.show()

# 3. Creating multiple figures
# Each call to plt.figure() creates a new, independent figure.
plt.figure(1) # Activates or creates figure 1
plt.plot([0, 1, 2], [0, 1, 4])
plt.title("First Figure")

plt.figure(2, figsize=(5, 5)) # Creates figure 2 with a custom size
plt.plot([0, 1, 2], [0, 2, 1])
plt.title("Second Figure")
plt.show()

# 4. Using 'num' to specify a figure number and 'clear' to clear it
# 'num' allows referencing figures by an integer or string ID.
# 'clear=True' clears the figure if it already exists before recreating it.
plt.figure(num='MyPlot', figsize=(7, 4), clear=True)
x = np.linspace(0, 5, 50)
plt.scatter(x, x**2)
plt.title("Figure with String ID and Cleared")
plt.show()

# 5. Setting DPI (Dots Per Inch) for higher resolution
plt.figure(dpi=150)
plt.plot(np.random.rand(20), 'o-')
plt.title("High Resolution Figure")
plt.show()

# 6. Creating a figure and adding subplots manually
# While plt.subplots() is often preferred, you can add axes manually.
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 2, 1) # 1 row, 2 columns, first subplot
ax1.plot([1, 2, 3], [5, 7, 6])
ax1.set_title("Subplot 1")

ax2 = fig.add_subplot(1, 2, 2) # 1 row, 2 columns, second subplot
ax2.bar(['A', 'B', 'C'], [10, 15, 8])
ax2.set_title("Subplot 2")
plt.tight_layout() # Adjusts subplot parameters for a tight layout
plt.show()
```
